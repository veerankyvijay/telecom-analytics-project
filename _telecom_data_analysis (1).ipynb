# Task 4 - Obtain the logical conditions for the features "Scr_h" and "Scr_w"

```python
# Let's tackle these features: "Scr_h", "Scr_w"
```

```python
# Create a new feature called "Scr_d" which stores the length of the diagonal of the screen of the phone

# Create new feature for diagonal
data["Scr_d"] = np.sqrt(data["Px_h"]**2 + data["Px_w"]**2)

```

```python
# Create a histogram of the "Scr_d" feature and also show the quartiles

q1 = data["Scr_d"].quantile(0.25)
q2 = data["Scr_d"].quantile(0.50)  # Median
q3 = data["Scr_d"].quantile(0.75)

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(data["Scr_d"], bins=10, edgecolor="black", alpha=0.7)

# Add quartile lines
plt.axvline(q1, color="red", linestyle="dashed", linewidth=2, label=f"Q1: {q1:.0f}")
plt.axvline(q2, color="green", linestyle="dotted", linewidth=2, label=f"Q2 (Median): {q2:.0f}")
plt.axvline(q3, color="blue", linestyle="dashdot", linewidth=2, label=f"Q3: {q3:.0f}")
plt.title("Histogram of Screen Diagonal (Scr_d)")
plt.xlabel("Diagonal (pixels)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Print quartiles
print("Q1 (25%):", q1)
print("Q2 (Median):", q2)
print("Q3 (75%):", q3)
```

```python
# The children want phones that have very good screen sizes
# Consider the phones that have screen sizes greater than or equal to the upper quartile value in the data set
# Create a logical condition for this situation and store the logical values as "con3"
# Create the 'Scr' feature, which is the total screen size
data["Scr"] = data["Scr_h"] * data["Scr_w"]

# Calculate the upper quartile of the 'Scr' feature
upper_quartile_scr = data["Scr"].quantile(0.75)

# Create the 'con3' logical condition
data["con3"] = data["Scr"] >= upper_quartile_scr

# Print the upper quartile screen size for reference
print(f"The upper quartile screen size (Scr) is: {upper_quartile_scr}")

# Print the value counts for the new condition
print("\nCounts of True and False in 'con3':")
print(data["con3"].value_counts())
```

Observations:

The features "Scr_h" and "Scr_w" are respectively the height and the width of the phone screen.

We created a new feature called "Scr_d" which is essentially the length of the screen diagonal.

The upper quartile has been selected as a threshold in this case as the children were very particular on this point.

In case it is too strict, we can choose the mean or the median as a threshold.

# Task 8 - Subset the data based on all the logical conditions

```python
# Subset the dataframe using all the logical conditions that have been stored
# Store the subset of the dataframe as a new dataframe called "df1"
# con2: Screen resolution (Px) >= median of Px
data["Px"] = data["Px_h"] * data["Px_w"]
data["con2"] = data["Px"] >= data["Px"].median()

# con3: Screen size (Scr) >= upper quartile of Scr
data["Scr"] = data["Scr_h"] * data["Scr_w"]
data["con3"] = data["Scr"] >= data["Scr"].quantile(0.75)

# con4: PC >= mean(PC) AND FC >= mean(FC)
data["con4"] = (data["PC"] >= data["PC"].mean()) & (data["FC"] >= data["FC"].mean())

# con5: Int_Mem >= mean(Int_Mem) AND Bty_Pwr >= mean(Bty_Pwr) AND RAM >= mean(RAM)
data["con5"] = (data["Int_Mem"] >= data["Int_Mem"].mean()) & (data["Bty_Pwr"] >= data["Bty_Pwr"].mean()) & (data["RAM"] >= data["RAM"].mean())

# con6: Depth <= median(Depth) AND Weight <= median(Weight)
data["con6"] = (data["Depth"] <= data["Depth"].median()) & (data["Weight"] <= data["Weight"].median())

# Combine all logical conditions to subset the DataFrame
df1 = data[data["con2"] & data["con3"] & data["con4"] & data["con5"] & data["con6"]]

# Display the first few rows of the new dataframe
print("First 5 rows of the new dataframe (df1):")
print(df1.head())

# Display the shape of the new dataframe
print("\nShape of the new dataframe (df1):")
print(df1.shape)


```

# Task 9 - Study the variability of the features in the original data set

```python
# Calculate the ratio of the standard deviation to the mean for all the numerical features in the dataframe
# Store these values in a new series wherein the rows are the features and the only column is the calculated ratio
# Name the series as "deviations"

numerical_features = [
    "Px_h",
    "Px_w",
    "Scr_h",
    "Scr_w",
    "PC",
    "FC",
    "Int_Mem",
    "Bty_Pwr",
    "RAM",
    "Depth",
    "Weight",
    "Price",
]
numerical_data = data[numerical_features]

# Calculate the standard deviation and mean of the numerical features
std_devs = numerical_data.std()
means = numerical_data.mean()

# Calculate the ratio of standard deviation to mean (Coefficient of Variation)
deviations = std_devs / means

# Name the series as "deviations"
deviations.name = "deviations"

# Print the resulting series
print(deviations)
```

```python
# View the "deviations" series after sorting it in descending order
numerical_features = [
    "Px_h",
    "Px_w",
    "Scr_h",
    "Scr_w",
    "PC",
    "FC",
    "Int_Mem",
    "Bty_Pwr",
    "RAM",
    "Depth",
    "Weight",
    "Price",
]
numerical_data = data[numerical_features]

# Calculate the standard deviation and mean of the numerical features
std_devs = numerical_data.std()
means = numerical_data.mean()

# Calculate the ratio of standard deviation to mean
deviations = std_devs / means

# Sort the "deviations" series in descending order
sorted_deviations = deviations.sort_values(ascending=False)

# Print the sorted series
print(sorted_deviations)

```

Observations:

The ratio of the standard deviation to the mean of a feature normalises it in a way.

This allows for comparison between multiple features.

The most variable feature in the original data set is the internal memory of the phones.

The least variable feature in the original data set is the number of screen pixels in the horizontal axis.

Although most features don't seem so variable, the prices of the phones are quite variable.
